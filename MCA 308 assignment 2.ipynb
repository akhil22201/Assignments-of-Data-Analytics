```python
# Q8
import pandas as pd

# Read CSV
df = pd.read_csv("employees.csv")

# a) Shape
print("Shape of DataFrame:", df.shape)

# b) Summary
print("\nSummary of DataFrame:")
print(df.info())

# c) Descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe())

# d) First 5 rows and last 3 rows
print("\nFirst 5 rows:")
print(df.head())
print("\nLast 3 rows:")
print(df.tail(3))

# e) Calculations
print("\nAverage Salary:", df["Salary"].mean())
print("Total Bonus:", df["Bonus"].sum())
print("Youngest Employee Age:", df["Age"].min())
print("Highest Rating:", df["Rating"].max())

# f) Sort by Salary
print("\nSorted by Salary (Descending):")
print(df.sort_values(by="Salary", ascending=False))

# g) Performance category
def categorize_performance(rating):
    if rating >= 4.5:
        return "Excellent"
    elif rating >= 4.0:
        return "Good"
    else:
        return "Average"

df["Performance_Category"] = df["Rating"].apply(categorize_performance)
print("\nDataFrame with Performance Category:")
print(df)

# h) Missing values
print("\nMissing Values in DataFrame:")
print(df.isnull().sum())

# i) Rename column
df.rename(columns={"Employee_ID": "ID"}, inplace=True)

# j) Filters
print("\nEmployees with more than 5 years of experience:")
print(df[df["Years_of_Experience"] > 5])
print("\nEmployees in IT Department:")
print(df[df["Department"] == "IT"])

# k) Add Tax column (10% of salary)
df["Tax"] = df["Salary"] * 0.10
print("\nDataFrame with Tax column:")
print(df)

# l) Save modified DataFrame
df.to_csv("employees_modified.csv", index=False)
print("\nModified DataFrame saved as employees_modified.csv")
```
