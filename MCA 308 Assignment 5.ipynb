```python
# Q1
import numpy as np
from scipy.spatial import distance

# Given customer rating vectors
customer_A = np.array([4, 5, 2, 3, 4])
customer_B = np.array([5, 3, 2, 4, 5])

# Binary preferences (1 = liked, 0 = disliked)
customer_A_binary = np.array([1, 0, 1, 1, 0, 1])
customer_B_binary = np.array([1, 1, 1, 0, 0, 1])

# Euclidean Distance – Overall difference in ratings
euclidean_dist = np.linalg.norm(customer_A - customer_B)
print(f"1. Euclidean Distance: {euclidean_dist:.4f}")

# Manhattan Distance – Total absolute deviation
manhattan_dist = np.sum(np.abs(customer_A - customer_B))
print(f"2. Manhattan Distance: {manhattan_dist:.4f}")

# Cosine Similarity – Alignment of customer interests
cosine_similarity = 1 - distance.cosine(customer_A, customer_B)
print(f"3. Cosine Similarity: {cosine_similarity:.4f}")

# Hamming Distance – Binary preference difference
hamming_dist = distance.hamming(customer_A_binary, customer_B_binary) * len(customer_A_binary)
print(f"4. Hamming Distance: {hamming_dist:.4f}")

# Jaccard Similarity – Common likes over total likes
intersection = np.logical_and(customer_A_binary, customer_B_binary).sum()
union = np.logical_or(customer_A_binary, customer_B_binary).sum()
jaccard_similarity = intersection / union
print(f"5. Jaccard Similarity: {jaccard_similarity:.4f}")
```
